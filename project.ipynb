# Classification

```python
df = pd.read_csv("boston.csv")
```

```python
y_class = pd.cut(df["MEDV"], bins=[0, 21.2, 25, 50], labels=["Low", "Medium", "High"])

x = df.drop("MEDV", axis=1)
y = y_class
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, Y_train, Y_test = train_test_split( x, y , test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy: {:.2f}".format(accuracy))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy: {:.2f}".format(accuracy))
```

Above, we train two models: RandomForest and DecisionTree. We tested the accuracy of these two models, and RandomForest achieved an accuracy of 0.8, while the DecisionTree achieved an accuracy of 0.71. Comparing the two, we can know that RandomForest performed better than DecisionTree. This is likely due to the RandomForeset using multiple decisions trees for classification, creating a better model in ways such as eliminating the issue of overfitting that a single decision tree often has.

# Regression

```python
df = pd.read_csv("boston.csv")
```

```python
x = df.drop("MEDV", axis=1)
y = df["MEDV"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

model = RandomForestRegressor(n_estimators=100, random_state=42)


model.fit(X_train, Y_train)


Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print('MSE:', mse)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print("Mean squared error:", mse)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(n_estimators=100, random_state=42)


model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print('MSE:', mse)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=42)


model.fit(X_train, Y_train)


Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
print('MSE:', mse)
```

For regression analysis and testing, we trained four different models. Out of these four models, the GradientBoostingRegressor had the lowest mean squared error and thus we can say that the GradientBoostingRegressor model performed better than the LinearRegresion, RandomForestRegression and the DecisionTreeRegression. This is likely due to the GradientBoostingRegressor model being able to fit the data better than the other models due to it being an ensemble method that can choose the better model by penalizing higher loss values, and thus having a lower mean squared error.

# Clustering

```python
df = pd.read_csv("boston.csv")
target = df["MEDV"]
df = df.drop("MEDV", axis=1)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0).fit(df)
clusters = kmeans.labels_ 
centroids = kmeans.cluster_centers_


plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='black')
plt.title('K-means Clustering')
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.show()


#target = df["MEDV"]
for i in range(len(centroids)):
    print('Cluster {}: Mean target value = {:.2f}'.format(i, np.mean(target[clusters == i])))
```

As we can see from the above graph, this is not the best example of clustering. The green and red dots overlap each other, which means that the model is having trouble distinguishing between the two. Let us perform some data cleaning to see if we can improve the model by scaling the data.

```python
#X = df.drop('MEDV', axis=1)
X=df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
# Elbow method to find optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

As we can see from the plot above, an 'elbow' emerges at k equals 3. This means that the optimal number of clusters is 3. This is likely due to the fact that the data is not very spread out, and thus the optimal number of clusters is not very high.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
# Visualize the clusters
plt.scatter(X_scaled[y_kmeans == 0, 0], X_scaled[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_scaled[y_kmeans == 1, 0], X_scaled[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_scaled[y_kmeans == 2, 0], X_scaled[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Boston Housing')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

The clusters are now more distinctly separate, as can be seen in the plot above.

```python
import pylab as pl
import matplotlib.cm as cm
from IPython import display
def plot_clusters(data, clusters, centroids, title):
    pl.gcf().clear()

    # Plotting Data Points
    if clusters is None:
        pl.scatter(data[:,0], data[:,1], s=10, c='steelblue')
        pl.scatter(centroids[:,0], centroids[:,1], marker='*', c='r', s=150)
    else:
    # Different color clusters
        k = len(centroids)
        for i in range(k):
            colors = cm.rainbow(np.linspace(0, 1, 3))
            cluster_data = data[clusters == i, :]
            pl.scatter(cluster_data[:,0], cluster_data[:,1], s=10, color=colors[i])
            pl.scatter(centroids[i,0], centroids[i,1], marker='*', color='k', s=200)
      
    pl.title(title)
    display.display(pl.gcf())
    display.clear_output(wait=True)
```

### PCA Analysis

```python
from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(df)

k_means = KMeans(n_clusters = 3)
cluster = k_means.fit_predict(Y_sklearn)
centroids = k_means.cluster_centers_
plot_clusters(Y_sklearn, cluster, centroids,"Clustering")
```
